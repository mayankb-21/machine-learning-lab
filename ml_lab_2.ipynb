```python
X = df.drop(['User ID','Purchased','Gender'], axis=1)
y = df['Purchased']

```

```python
train_size = int(0.8 * len(df))

X_train = X.iloc[:train_size]
y_train= y.iloc[:train_size]

X_test = X.iloc[train_size:]
y_test = y.iloc[train_size:]
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
class LogisticRegression():

    def __init__(self, lr, n_iters,lambda_param):
        self.lr = lr
        self.n_iters = n_iters
        self.lambda_param = lambda_param
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y).reshape(-1, 1)
        n_samples, n_features = X.shape
        self.weights = np.zeros((n_features, 1))
        self.bias = 0

        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            predictions = sigmoid(linear_pred)
            dw = (1/n_samples) * np.dot(X.T, (predictions - y)) + (self.lambda_param / n_samples) * self.weights
            db = (1/n_samples) * np.sum(predictions-y)

    def predict(self, X):
        X = np.array(X)
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        class_pred = [0 if y<=0.5 else 1 for y in y_pred]
        return class_pred

```

```python
clf = LogisticRegression(lr=0.01, n_iters=5000,lambda_param=0.1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
def accuracy (y_pred, y_test):
    return np.sum(y_pred==y_test)/len(y_test)
```

```python
acc = accuracy(y_pred, y_test)
acc
```
